1. Importación de los datos

```python
import pandas as pd
df = pd.read_csv('[CHANGE THIS PATH]/df_iob.csv', encoding = "UTF-8")
df = df.drop(['Unnamed: 0'], axis=1)
df.tail()
```

2. Creación del diccionario para almacenar los experimentos

```python
dict_count = {
    'document':[],
    'wcount':[], 
    'type':[],
    'category':[]
    }

list_tag_unique = pd.unique(df['Tag'])

for t in list_tag_unique:
    if t.startswith('B'):
        dict_count[t] = []

dict_count
```

3. Generación del DF

```python

for ind in df.index:
    if df['MD5'][ind] not in dict_count['document']:
        dict_count['document'].append(df['MD5'][ind])
        index_doc = dict_count['document'].index(df['MD5'][ind])
        dict_count['type'].append(df['type'][ind])
        dict_count['category'].append(df['category'][ind])
        for k in dict_count:
            if k.startswith('B-'):
                dict_count[k].append(0)
        if df['Tag'][ind].startswith('B-'):
            dict_count[df['Tag'][ind]][index_doc] += 1
        init_wcount = ind

    if df['MD5'][ind] in dict_count['document']:
        index_doc = dict_count['document'].index(df['MD5'][ind])
        if df['Tag'][ind].startswith('B-'):
            dict_count[df['Tag'][ind]][index_doc] += 1
        if ind == len(df)-1:
            end_wcount = ind
            dict_count['wcount'].append(end_wcount-init_wcount)
        elif df['MD5'][ind+1] != df['MD5'][ind]:
            print(f'''is {df['MD5'][ind]} actual line and {df['MD5'][ind+1]}''')
            end_wcount = ind
            dict_count['wcount'].append(end_wcount-init_wcount)
    
df_count = pd.DataFrame.from_dict(dict_count)
```

5. Mostrar los tags más frecuentes

```python
df_count.dtypes
```

```python
test_tag = {}
for c in df_count.columns:
    if c.startswith('B-'):
        tot = df_count[c].sum()
        test_tag[c] = tot


tags2keep = []
for k in test_tag:
    if test_tag[k] > 400:
        tags2keep.append(k)

tags2keep
```

```python
for c in df_count.columns:
    if c.startswith('B-'):
        if c not in tags2keep:
            df_count = df_count.drop(c, axis=1)
```

```python
df_count['type'].value_counts()

for ind in df_count.index:
    if df_count['type'][ind] not in ['Contestación de demanda', 'Manifestación de parte', 'Solicitud de copias certificadas', 'Consignación billete de deposito']:
        df_count = df_count.drop(ind)

df_count = df_count.reset_index(drop=True)
```
